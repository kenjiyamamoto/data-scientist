```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Functional
plt.plot(x,y,'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y)
axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('Título')
axes2.plot(y,x)
```
